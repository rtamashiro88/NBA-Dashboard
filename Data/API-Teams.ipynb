```python
from API import KEY, URL
from pandas import DataFrame
from requests import get
from json import dumps
```

```python
# API 'GET' Request
qry = f'{URL}teams/NBA/?key={KEY}&season=2024&max=1000'
data = get(qry).json()
data.keys()
```

```python
print(dumps(data, indent=4))
```

```python
print(*data['teams'].values(), sep='\n')
```

```python
df = DataFrame(data=data['teams'].values())
```

```python
df.head()
```

```python
df = df.rename(columns=
    {
    'ID': 'TeamID',
    'Name': 'Description'
    }
)
```

```python
df.dtypes
```

```python
df['TeamID'] = df['TeamID'].astype(int)
df['Active'] = df['Active'].astype(bool)
```

```python
df.dtypes
```

```python
team_data = df \
    .set_index('TeamID') \
    .sort_index()
```

```python
team_data['Name'] = team_data['Description'].apply(lambda x: x.split(" ")[1].strip())
team_data['City'] = team_data['Description'].apply(lambda x: x.split(" ")[0].strip())
team_data.head()
```

```python
team_data = team_data[
    [
    'Name',
    'City',
    'Abbrev',
    'Description',
    'Active'
    ]
]
```
