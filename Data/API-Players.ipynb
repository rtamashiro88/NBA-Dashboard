```python
from API import KEY, URL
from pandas import DataFrame
from requests import get
from json import dumps
```

```python
# API 'GET' Request
qry = f'{URL}players/NBA/?key={KEY}&season=2024&max=1000'
data = get(qry).json()
data.keys()
```

```python
print(dumps(data['players'], indent=4))
```

```python
current_teams = list()
for k, v in data['players'].items():
    if type(v) is list:
        current_teams.append(v[0])
    else:
        current_teams.append(v)
```
